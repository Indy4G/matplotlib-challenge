# 3. Generally, the more a mouse weighs, the smaller the amount of tumor volume loss will be over time.

## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset
full_study = mouse_metadata.merge(study_results, on="Mouse ID")
```

## Summary statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
stats_df = full_study[["Drug Regimen", "Tumor Volume (mm3)"]]
mean = stats_df.groupby(["Drug Regimen"]).mean()
median = stats_df.groupby(["Drug Regimen"]).median()
variance = stats_df.groupby(["Drug Regimen"]).var()
std_dev = stats_df.groupby(["Drug Regimen"]).std()
sem = stats_df.groupby(["Drug Regimen"]).sem()
stats_df = pd.DataFrame({
    "Mean":mean["Tumor Volume (mm3)"],
    "Median":median["Tumor Volume (mm3)"],
    "Variance":variance["Tumor Volume (mm3)"],
    "Standard Deviation":std_dev["Tumor Volume (mm3)"],
    "SEM":sem["Tumor Volume (mm3)"]
})
stats_df
```

## Bar plots

```python
# Generate a bar plot showing number of data points for each treatment regimen using pandas
data_points_df = full_study[["Mouse ID","Drug Regimen"]]
data_points = data_points_df.groupby(["Drug Regimen"]).count()
data_points = data_points.rename(columns={"Mouse ID":"Data Points"})
data_points.plot(kind='bar',legend=False,title=("Data Points per Drug Regimen"));
```

```python
# Generate a bar plot showing number of data points for each treatment regimen using pyplot
data_points = data_points.reset_index(drop=True)
x_axis = np.arange(len(data_points))
y_axis = data_points["Data Points"].tolist()
x_ticks = data_points["Drug Regimen"].tolist()
plt.bar(x_axis,y_axis)
plt.xticks(x_axis, x_ticks, rotation='vertical')
plt.title("Data Points per Drug Regimen")
plt.xlabel("Drug Regimen")
plt.tight_layout()
plt.show();
```

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens: 
#Capomulin, Ramicane, Infubinol, and Ceftamin. Calculate the IQR and quantitatively determine if there 
#are any potential outliers.
final_vol_all = full_study[["Mouse ID","Timepoint","Tumor Volume (mm3)","Drug Regimen"]]
final_vol_all = final_vol_all.drop_duplicates(subset="Mouse ID", keep="last")

Cap = final_vol_all[final_vol_all['Drug Regimen'] == 'Capomulin']
Cap_Q1 = Cap['Tumor Volume (mm3)'].quantile(0.25)
Cap_Q3 = Cap['Tumor Volume (mm3)'].quantile(0.75)
Cap_IQR = Cap_Q3 - Cap_Q1
Cap_lowerOuts = Cap[Cap['Tumor Volume (mm3)'] < (Cap_Q1-(1.5*Cap_IQR))]
Cap_upperOuts = Cap[Cap['Tumor Volume (mm3)'] > (Cap_Q3+(1.5*Cap_IQR))]
Cap_Outs = pd.concat([Cap_lowerOuts,Cap_upperOuts],axis=0)
print(f'There are {len(Cap_Outs)} potential outlier(s) in the Capomulin Regimen mice. The IQR was {"{:.2f}".format(Cap_IQR)}.')

Ram = final_vol_all[final_vol_all['Drug Regimen'] == 'Ramicane']
Ram_Q1 = Ram['Tumor Volume (mm3)'].quantile(0.25)
Ram_Q3 = Ram['Tumor Volume (mm3)'].quantile(0.75)
Ram_IQR = Ram_Q3 - Ram_Q1
Ram_lowerOuts = Ram[Ram['Tumor Volume (mm3)'] < (Ram_Q1-(1.5*Ram_IQR))]
Ram_upperOuts = Ram[Ram['Tumor Volume (mm3)'] > (Ram_Q3+(1.5*Ram_IQR))]
Ram_Outs = pd.concat([Ram_lowerOuts,Ram_upperOuts],axis=0)
print(f'There are {len(Ram_Outs)} potential outlier(s) in the Ramicane Regimen mice. The IQR was {"{:.2f}".format(Ram_IQR)}.')

Inf = final_vol_all[final_vol_all['Drug Regimen'] == 'Infubinol']
Inf_Q1 = Inf['Tumor Volume (mm3)'].quantile(0.25)
Inf_Q3 = Inf['Tumor Volume (mm3)'].quantile(0.75)
Inf_IQR = Inf_Q3 - Inf_Q1
Inf_lowerOuts = Inf[Inf['Tumor Volume (mm3)'] < (Inf_Q1-(1.5*Inf_IQR))]
Inf_upperOuts = Inf[Inf['Tumor Volume (mm3)'] > (Inf_Q3+(1.5*Inf_IQR))]
Inf_Outs = pd.concat([Inf_lowerOuts,Inf_upperOuts],axis=0)
print(f'There are {len(Inf_Outs)} potential outlier(s) in the Infubinol Regimen mice. The IQR was {"{:.2f}".format(Inf_IQR)}.')

Cef = final_vol_all[final_vol_all['Drug Regimen'] == 'Ceftamin']
Cef_Q1 = Cef['Tumor Volume (mm3)'].quantile(0.25)
Cef_Q3 = Cef['Tumor Volume (mm3)'].quantile(0.75)
Cef_IQR = Cef_Q3 - Cef_Q1
Cef_lowerOuts = Cef[Cef['Tumor Volume (mm3)'] < (Cef_Q1-(1.5*Cef_IQR))]
Cef_upperOuts = Cef[Cef['Tumor Volume (mm3)'] > (Cef_Q3+(1.5*Cef_IQR))]
Cef_Outs = pd.concat([Cef_lowerOuts,Cef_upperOuts],axis=0)
print(f'There are {len(Cef_Outs)} potential outlier(s) in the Ceftamin Regimen mice. The IQR was {"{:.2f}".format(Cef_IQR)}.')
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
Tumor_Vols = [Cap['Tumor Volume (mm3)'],
              Ram['Tumor Volume (mm3)'],
              Inf['Tumor Volume (mm3)'],
              Cef['Tumor Volume (mm3)']]

plt.boxplot(Tumor_Vols)
plt.title('Final Tumor Volumes Across the Four Regimens of Interest')
x_axis = [1,2,3,4]
x_ticks = ["Capomulin","Ramicane","Infubinol","Ceftamin"]
plt.xticks(x_axis, x_ticks)
plt.xlabel("Drug Regimen")
plt.ylabel("Final Tumor Volume in mm3")
plt.show()
```

## Line and scatter plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
weightandvol = full_study[["Mouse ID","Timepoint","Tumor Volume (mm3)","Drug Regimen","Weight (g)"]]
Cap_weightandvol= weightandvol[weightandvol['Drug Regimen'] == 'Capomulin']
Cap_weightandvol_s185 = Cap_weightandvol[Cap_weightandvol['Mouse ID'] == 's185']
time = Cap_weightandvol_s185["Timepoint"].tolist()
tumor_vol = Cap_weightandvol_s185["Tumor Volume (mm3)"].tolist()
plt.plot(time,tumor_vol)
plt.title("Tumor Volume over Time for Mouse s185")
plt.xlabel("Time")
plt.ylabel("Tumor Volume in mm3")
plt.xlim(0,45)
plt.ylim(20,50)
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
avg_tumor_vol_df = Cap_weightandvol.groupby(["Tumor Volume (mm3)"]).mean()
avg_tumor_vol_df = avg_tumor_vol_df.reset_index()
avg_tumor_vol_df = avg_tumor_vol_df.rename({"Tumor Volume (mm3)":"Average Tumor Volume (mm3)"})
weights = avg_tumor_vol_df["Weight (g)"].tolist()
avg_tumor_vol = avg_tumor_vol_df["Tumor Volume (mm3)"].tolist()

plt.scatter(weights,avg_tumor_vol)
plt.title("Mouse Weight vs. Average Tumor Volume")
plt.xlabel("Mouse Weight")
plt.ylabel("Average Tumor Volume in mm3")
plt.xlim(14.9,25.1)
plt.show();
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and 
#average tumor volume for the Capomulin regimen
(slope, intercept, rvalue, _, _) = st.linregress(weights,avg_tumor_vol)

plt.scatter(weights,avg_tumor_vol)
plt.title("Mouse Weight vs. Average Tumor Volume")
plt.xlabel("Mouse Weight")
plt.ylabel("Average Tumor Volume in mm3")
plt.xlim(14.9,25.1)
weights = [int(i) for i in weights]
regress_vals = [i * int(slope) + int(intercept) for i in weights]
plt.plot(weights,regress_vals)
line_eq = f'y = {"{:.2f}".format(slope)}*x + {"{:.2f}".format(intercept)}'
plt.annotate(line_eq, (20,30), fontsize=15, color='r')
plt.show();
print(f'The Correlation Coefficient is {"{:.2f}".format(rvalue)}')
```
